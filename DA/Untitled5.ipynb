```python
import pandas as pd
import numpy as np
df = pd.read_csv("OnlineRetail.csv", encoding = "ISO-8859-1")
df
```

```python
# Add a new column with the number of items for each row
df['Num Items'] = df['Quantity'] * df['UnitPrice']

# Build a pivot table with the average number of items per invoice by country
pivot_table = pd.pivot_table(df, values='Num Items', index='InvoiceNo',
                             columns='Country', aggfunc='mean')

# Print the result
print(pivot_table)
```

```python
pivot_table = pd.pivot_table(df, values='Quantity', index=['CustomerID', 'StockCode'],
                             aggfunc=[min, max])

print(pivot_table)
```

```python
df['Total Cost'] = df['Quantity'] * df['UnitPrice']

# Build a pivot table with the total number of items and the average price by warehouse code
pivot_table = pd.pivot_table(df, values=['Quantity', 'Total Cost'], index='StockCode',
                             aggfunc={'Quantity': 'sum', 'Total Cost': 'mean'})

print(pivot_table)
```
