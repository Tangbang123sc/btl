```python
#Ph√©p to√°n t·ª≠ so s√°nh: ==, !=, >, < ,>=,<= 
#Ph√©p to√°n logic: and &, or | , not ~ 
#L·ªçc d·ªØ li·ªáu v·ªõi query 
#df.query("(T√™n_c·ªôt_1 PTSS gi√°_tr·ªã_1) PTLG (T√™n_c·ªôt_n PTSS gi√°_tr·ªã_n)")
#Query n·∫øu ch√∫ng ta define gi√° tr·ªã l√† object d·∫°ng chu·ªói, th√¨ ph·∫£i ƒë·ªÉ d·∫•u nh√°y ƒë∆°n '' thay v√¨ d·∫•u nh√°y k√©p ƒë·ªÉ Python ph√¢n bi·ªát
df.query("(name == 'R&O RedOrange')") #S·ª≠a d·∫•u nh√°y c·ªßa RedOrgange th√†nh nh√°y k√©p ƒë·ªÉ th·∫•y s·ª± kh√°c bi·ªát

#H√£y l·∫•y ra nh·ªØng shop c√≥ (th√°ng tham gia(join_month) == September ho·∫∑c item_count> 200) v√† follower_count > 500000
#ƒê·ªëi v·ªõi query, t·ªïng t·∫•t c·∫£ nh·ªØng ƒëi·ªÅu ki·ªán ƒë·ªÅu ƒë∆∞·ª£c b·ªçc trong d·∫•u ngo·∫∑c k√©p ""
df.query("((join_month == 'September') | (item_count >200)) & (follower_count>500000)")
```

```python
#GROUPBY
#dataFrame.groupby(‚ÄòThu·ªôc t√≠nh c·∫ßn ph√¢n nh√≥m‚Äô)[‚ÄòThu·ªôc t√≠nh c·∫ßn t√≠nh gi√° tr·ªã‚Äô].ph√©p to√°n()
#T√≠nh t·ªïng s·ªë l∆∞·ª£ng rating_bad, rating_good, v√† follower_count c·ªßa t·∫•t c·∫£ c√°c chi nh√°nh cho t·ª´ng t√™n shop
df.groupby("name")["rating_good","rating_bad","follower_count"].sum().sort_values("follower_count", ascending = False)
#sort_values("s·∫Øp x·∫øp tr√™n c·ªôt n√†o",ascending = False c√≥ nghƒ©a l√† kh√¥ng tƒÉng d·∫ßn)
df["LuongTuongTac"] = df["rating_bad"] + df["rating_good"] + df["rating_normal"]
df[df["LuongTuongTac"] >1000000][["name","follower_count","LuongTuongTac"]].sort_values("LuongTuongTac",ascending = False)
```

```python
#T·∫°o m·ªôt c·ªôt d·ª±a tr√™n ƒëi·ªÅu ki·ªán s·ª≠ d·ª•ng numpy.where
#df['rate'] = np.where(df['rating_good'] >= 50000,'good','bad' )
#N·∫øu m·ªôt c·ª≠a h√†ng c√≥ rating_good > 500000 v√† follower_count > 1000000 th√¨ h·ªç ƒë∆∞·ª£c x·∫øp lo·∫°i Michellin, c√≤n l·∫°i x·∫øp lo·∫°i b√¨nh th∆∞·ªùng
df["XepLoai"] = np.where((df["rating_good"]>500000) & (df["follower_count"]>1000000),"Michellin","BinhThuong")
df
```

```python
#Th√™m c·ªôt m·ªõi theo multiple conditions(nhi·ªÅu ƒëi·ªÅu ki·ªán)
#V√≠ d·ª•: Th√™m c·ªôt flag t·∫∑ng c·ªù cho c√°c c·ª≠a h√†ng, flag nh·∫≠n c√°c gi√° tr·ªã nh∆∞ sau:
# 	blue khi rating_good >= 30000 v√† rating_bad <= 100
# 	yellow khi 10000 <= rating_good < 30000 v√† 100 < rating_bad <= 1000
# 	red khi rating_good < 10000
#	black ƒë·ªëi v·ªõi c√°c tr∆∞·ªùng h·ª£p c√≤n l·∫°i.
#C√°ch th·ª±c thi chu·∫©n khi d√πng np.select()
conditions = [(df["rating_good"] >= 300000) & (df["rating_bad"]<= 100),
             (df["rating_good"]<30000) & (10000<= df["rating_good"]) &(100<df["rating_bad"]) & (df["rating_bad"]<=1000),
             (df["rating_good"]<10000)]
choices = ["blue","yellow","red"]
df["flag"] = np.select(conditions,choices,default = "black")
df
```
