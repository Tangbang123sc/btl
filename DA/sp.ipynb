```python
#Phép toán tử so sánh: ==, !=, >, < ,>=,<= 
#Phép toán logic: and &, or | , not ~ 
#Lọc dữ liệu với query 
#df.query("(Tên_cột_1 PTSS giá_trị_1) PTLG (Tên_cột_n PTSS giá_trị_n)")
#Query nếu chúng ta define giá trị là object dạng chuỗi, thì phải để dấu nháy đơn '' thay vì dấu nháy kép để Python phân biệt
df.query("(name == 'R&O RedOrange')") #Sửa dấu nháy của RedOrgange thành nháy kép để thấy sự khác biệt

#Hãy lấy ra những shop có (tháng tham gia(join_month) == September hoặc item_count> 200) và follower_count > 500000
#Đối với query, tổng tất cả những điều kiện đều được bọc trong dấu ngoặc kép ""
df.query("((join_month == 'September') | (item_count >200)) & (follower_count>500000)")
```

```python
#GROUPBY
#dataFrame.groupby(‘Thuộc tính cần phân nhóm’)[‘Thuộc tính cần tính giá trị’].phép toán()
#Tính tổng số lượng rating_bad, rating_good, và follower_count của tất cả các chi nhánh cho từng tên shop
df.groupby("name")["rating_good","rating_bad","follower_count"].sum().sort_values("follower_count", ascending = False)
#sort_values("sắp xếp trên cột nào",ascending = False có nghĩa là không tăng dần)
df["LuongTuongTac"] = df["rating_bad"] + df["rating_good"] + df["rating_normal"]
df[df["LuongTuongTac"] >1000000][["name","follower_count","LuongTuongTac"]].sort_values("LuongTuongTac",ascending = False)
```

```python
#Tạo một cột dựa trên điều kiện sử dụng numpy.where
#df['rate'] = np.where(df['rating_good'] >= 50000,'good','bad' )
#Nếu một cửa hàng có rating_good > 500000 và follower_count > 1000000 thì họ được xếp loại Michellin, còn lại xếp loại bình thường
df["XepLoai"] = np.where((df["rating_good"]>500000) & (df["follower_count"]>1000000),"Michellin","BinhThuong")
df
```

```python
#Thêm cột mới theo multiple conditions(nhiều điều kiện)
#Ví dụ: Thêm cột flag tặng cờ cho các cửa hàng, flag nhận các giá trị như sau:
# 	blue khi rating_good >= 30000 và rating_bad <= 100
# 	yellow khi 10000 <= rating_good < 30000 và 100 < rating_bad <= 1000
# 	red khi rating_good < 10000
#	black đối với các trường hợp còn lại.
#Cách thực thi chuẩn khi dùng np.select()
conditions = [(df["rating_good"] >= 300000) & (df["rating_bad"]<= 100),
             (df["rating_good"]<30000) & (10000<= df["rating_good"]) &(100<df["rating_bad"]) & (df["rating_bad"]<=1000),
             (df["rating_good"]<10000)]
choices = ["blue","yellow","red"]
df["flag"] = np.select(conditions,choices,default = "black")
df
```
