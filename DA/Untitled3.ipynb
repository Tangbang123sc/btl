```python
import matplotlib.pyplot as plt

# data for house price vs area
area = [1200, 1400, 1600, 1800, 2000, 2200]
price = [250000, 280000, 300000, 330000, 360000, 400000]

# plot house price vs area
plt.plot(area, price, marker='o')
plt.xlabel('Area (sqft)')
plt.ylabel('Price ($)')
plt.title('House Price vs Area')
plt.show()

# data for house price vs number of bedrooms
bedrooms = [2, 3, 4, 5, 6]
price = [200000, 250000, 300000, 350000, 400000]

# plot house price vs number of bedrooms
plt.plot(bedrooms, price, marker='o')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price ($)')
plt.title('House Price vs Number of Bedrooms')
plt.show()
```
